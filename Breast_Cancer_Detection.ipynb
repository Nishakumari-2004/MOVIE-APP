```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
#adding the target column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
#separating the features and target
X= data_frame.drop(columns='label', axis=1) #features,dropped label
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
  #standardising the data
  from sklearn.preprocessing import StandardScaler
  scaler = StandardScaler()
  X_train_std = scaler.fit_transform(X_train)
  X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

```python
import tensorflow as tf
tf.random.set_seed(3)  #to get same output
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),#converts data into 1 dimensional array
    keras.layers.Dense(20, activation='relu'),#number of neurons ,activation function
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
model .compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
#training the model
history = model.fit(X_train, Y_train, validation_split=0.1, epochs=10)
```

```python
#visualising accuracy and loss
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data', 'validation data'], loc='lower right')
plt.show()
```

```python
 plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data', 'validation data'], loc='upper right')
plt.show()
```

Accuracy of the model on test data


```python
loss,accuracy=model.evaluate(X_test,Y_test)
print(accuracy)
```
